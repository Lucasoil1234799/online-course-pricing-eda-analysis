```python
course_name=[]
Platform=[]
Rating=[]
reviews=[]
course_type=[]
#queries=["Data Analyst"]
for i in range(1,10):
    url1=f'https://www.classcentral.com/search?q=data+analyst&page={i}'
    url2=f'https://www.classcentral.com/search?q=data+science&page={i}'
    url3=f'https://www.classcentral.com/search?q=data+engineering&page={i}'
    for url in [url1,url2,url3]:
        elements=driver.get(url)
        for elements in driver.find_elements(By.TAG_NAME,'h2'):
            text=elements.text.strip()
            course_name.append(text)
        for elements in driver.find_elements(By.CSS_SELECTOR,".hover-underline.color-charcoal.text-3.margin-left-small.line-tight"):
            text=elements.text.strip()
            Platform.append(text)
        for elements in driver.find_elements(By.CSS_SELECTOR,"span.cmpt-rating-medium"):
            text=elements.get_attribute("aria-label").strip()
            Rating.append(text)
        for elements in driver.find_elements(By.CSS_SELECTOR,'.text-3.color-gray.margin-left-xxsmall'):
            text=elements.text.strip()
            reviews.extend(re.findall('\d+',text))
        for elements in driver.find_elements(By.CSS_SELECTOR,'span[aria-label="Pricing"]'):
            text=elements.text.strip()
            course_type.append(text)
```

```python
from itertools import zip_longest
import numpy as np
```

```python
frame=list(zip_longest(course_name,Platform,Rating,reviews,course_type,fillvalue=np.nan))
```

```python
df=pd.DataFrame(frame,columns=["Course_Name","Platfrom","Rating","Reviews","Course_Type"])
df
```

```python
df.to_csv("courses.csv",index=False)
```
