## Evaluation (Hypothesis Testing)

** Null Hypothesis (H0)="There is no significant difference in average ratings between Free and Paid courses"

** Alternative Hypothesis H(A)="There is a significant difference in average ratings between Free and Paid courses"

```python
from scipy.stats import ttest_ind
paid=df[df["Course_Type"]=="Paid"]["Rating"]
free=df[df["Course_Type"]=="Free"]["Rating"]
t_static,p_value=ttest_ind(paid,free,equal_var=False)
print(t_static,p_value)
if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Falied to Reject Null Hypothesis")
```

** Is there a significant relationship between Reviews and Rating?

```python
from scipy.stats import pearsonr
```

```python
static,p_value=pearsonr(df["Rating"],df["Reviews"])
print(static,p_value)
if p_value < 0.05:
    print("Significant correlation")
else:
    print("Not Significant correlation")
```
